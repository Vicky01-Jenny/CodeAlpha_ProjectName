# Import Libraries

````python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots visually appealing
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 6)
````

# Load Dataset

````python
# Load Dataset
# Upload your Excel file from Colab sidebar or Google Drive
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_excel("/content/drive/MyDrive/CodeAlpha/Sample-Superstore.xls")

# Preview the dataset
df.head()
````

# Check column names

````python
# Check column names
print("\nColumn names:\n", df.columns)
````

# Explore and Clean the Data

````python
# Explore and Clean the Data

# Check structure
df.info()

# Check for missing values
print("\nMissing values per column:\n", df.isnull().sum())

# Drop duplicates
df.drop_duplicates(inplace=True)

# Drop missing values (if any)
df.dropna(inplace=True)

# Verify dataset shape
print("\nDataset Shape after cleaning:", df.shape)
````

# Basic Statistical Summary

````python
# Basic Statistical Summary
df.describe()
````

# Sales and Profit Distribution

```
# This is formatted as code
```



````python
# Sales and Profit Distribution
df(['Sales', 'Profit']).describe
plt.figure(figsize=(10,5))
sns.histplot(df['Sales'], kde=True, color='blue')
plt.title('Sales Distribution')
plt.show()

plt.figure(figsize=(10,5))
sns.histplot(df['Profit'], kde=True, color='green')
plt.title('Profit Distribution')
plt.show()
print('Sales are positively skewed â€” most transactions are small, few are large.')
````

# Relationship: Sales vs Profit

````python
# Relationship: Sales vs Profit


plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Category', alpha=0.8)
plt.title("Sales vs Profit by Category")
plt.show()
print('Some high sales bring negative profit â€” possibly due to excessive discounts.')
````

#Profit by Category

````python
#Profit by Category

plt.figure(figsize=(8,6))
sns.barplot(data=df, x='Category', y='Profit', estimator=sum, palette='Set2')
plt.title("Total Profit by Category")
plt.show()
print("Technology is the most profitable category; Furniture often lags behind.")
````

#Sales by Region

````python
#Sales by Region

plt.figure(figsize=(8,6))
sns.barplot(data=df, x='Region', y='Sales', estimator=sum, palette='coolwarm')
plt.title("Total Sales by Region")
plt.show()
print("Western and Eastern regions dominate sales.")
````

# Top 10 Profitable States

````python
# Top 10 Profitable States

top_states = df.groupby('State')['Profit'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_states.values, y=top_states.index, palette="mako")
plt.title("Top 10 Profitable States")
plt.xlabel("Profit")
plt.ylabel("State")
plt.show()
print("California and New York are the most profitable states.")
````

#Impact of Discount on Profit

````python
#Impact of Discount on Profit

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Discount', y='Profit', alpha=0.7, color='red')
plt.title("Effect of Discount on Profit")
plt.show()
print("Higher discounts have a negative impact on profit.")
````

#Sales Share by Category (Pie Chart)

````python
#Sales Share by Category (Pie Chart)

sales_by_category = df.groupby('Category')['Sales'].sum()
plt.figure(figsize=(6,6))
plt.pie(sales_by_category, labels=sales_by_category.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
plt.title("Sales Distribution by Category")
plt.show()
print("Technology dominate sales.")
````

#Correlation Heatmap

````python
#Correlation Heatmap

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
print("Sales and Profit are moderately correlated; Discount is negatively correlated with Profit.")
````

# Multi-Chart Dashboard

````python
# Multi-Chart Dashboard


fig, axes = plt.subplots(2, 2, figsize=(14,10))

sns.barplot(data=df, x='Category', y='Profit', estimator=sum, ax=axes[0,0],palette='coolwarm')
axes[0,0].set_title("Profit by Category")

sns.barplot(data=df, x='Region', y='Sales', estimator=sum, ax=axes[0,1],palette='coolwarm')
axes[0,1].set_title("Sales by Region")

sns.scatterplot(data=df, x='Discount', y='Profit', ax=axes[1,0],color='red')
axes[1,0].set_title("Discount vs Profit")

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", ax=axes[1,1])
axes[1,1].set_title("Correlation Heatmap")

plt.tight_layout()
plt.show()
print("This dashboard view summarizes key business patterns at once.")
````
